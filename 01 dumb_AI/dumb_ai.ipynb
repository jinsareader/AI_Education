```python
import numpy

#1. 입력값 x, 정답 t를 입력해요
x = numpy.array([[1],[2],[3],[4],[5]]);
t = numpy.array([[44],[47],[51],[56],[60]]);

#2. 함수 F(x)를 만들어요
## F(x) = x*W + b
W = numpy.array([[numpy.random.randn()]]);
b = numpy.array([[numpy.random.randn()]]);
def F(x) :
    return numpy.dot(x,W) + b;

for epoch in range(0,1000) :
    #3. 출력값 y를 출력해요
    y = F(x);
    
    #4. 출력값 y와 정답 t를 비교해요
    ##MSE 사용
    MSE = numpy.mean((y - t) ** 2);
    
    #5. 비교값을 이용해 F(x)를 수정해요
    #함수 수정할때 학습률(learning rate)를 이용해 함수를 수정하는 정도는 보정
    ##경사하강법(미분을 통해 함수를 수정) 사용
    
    Wd = numpy.mean(2 * (y - t) * x); #미분값
    bd = numpy.mean(2 * (y - t) * 1); #미분값
    lr = 0.01 ##학습률
    
    W = W - Wd * lr;
    b = b - bd * lr;

    print("epoch {} - loss {}".format(epoch, MSE));

#반복해요
```

```python
import matplotlib.pyplot as plt

flg, ax = plt.subplots();
ax.scatter(x,t);
ax.plot(x,y);
plt.show();
```
