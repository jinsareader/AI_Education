```python
import numpy
import torch
import torch.nn as nn

#1. 입력값 x, 정답 t를 입력해요
x = numpy.array([[1],[2],[3],[4],[5]]);
t = numpy.array([[44],[47],[51],[56],[60]]);
#torch는 torch.tensor형태의 데이터만 읽을 수 있어요.
tensor_x = torch.tensor(x); ##numpy 행렬을 tensor형태로 바꾸는 방법
tensor_t = torch.tensor([[44],[47],[51],[56],[60]], dtype = torch.float); ##list를 tensor형태로 바꾸는 방법
tensor_x = tensor_x.type(torch.float); ##형변환도 수동으로 해줘야 함

#2. 함수들(F(x) (레이어), y와 t를 비교하는 함수(손실함수), F(x)를 수정하는 함수(최적화 함수))을 만들어요
## F(x) = x*W + b
F = nn.Sequential(
    nn.Linear(1,1)
); #F(x)
loss_function = nn.MSELoss(); #손실함수
optimizer = torch.optim.SGD(F.parameters(), lr = 0.01); #최적화 함수

for epoch in range(0,1000) :
    #3. 출력값 y를 출력해요
    tensor_y = F(tensor_x);
    
    #4. 출력값 y와 정답 t를 비교해요
    ##MSE 사용
    MSE = loss_function(tensor_y, tensor_t);
    
    #5. 비교값을 이용해 F(x)를 수정해요
    #함수 수정할때 학습률(learning rate)를 이용해 함수를 수정하는 정도는 보정
    ##경사하강법(미분을 통해 함수를 수정) 사용
    MSE.backward();
    optimizer.step();
    optimizer.zero_grad();

    print("epoch {} - loss {}".format(epoch, MSE));

#반복해요
```

```python
torch.save(F,"test.pt")
```
