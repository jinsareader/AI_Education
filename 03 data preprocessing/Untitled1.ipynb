```python
data = pd.read_csv("titanic.csv")

display(data)

threshold = 0.6
# 1. 결측치가 너무 많은 열 제거
data = data[data.columns[data.isnull().mean() < threshold]]
#df.columns : 해당 df의 컬럼들을 리스트 형태로 불러오는 프로퍼티
#isnull은 비어있는 칸 갯수를 세어주는 함수

# 2. 결측치가 너무 많은 행 제거
data = data.loc[data.isnull().mean(axis=1) < (0.01)]

display(data)
```

```python
data = pd.read_csv("titanic.csv")

data = data[["Age","Fare"]]

## 결측치 치환
### data = data.fillna(0) ## 0으로 바꾸기
### data = data.fillna(data.mean(axis = 0)) ## 평균치로 바꾸기
data = data.fillna(data.median()) ## 중앙값으로 바꾸기

display(data)
```

```python
# 이상치 관리 (이상한 값 삭제)

data = pd.read_csv("titanic.csv")

display(data)
#요금값에 대한 이상치 삭제
factor = 2 ## 임의로 정하는 보정치 (큰 의미 없음)
max_ = data["Fare"].mean() + data["Fare"].std() * factor
min_ = data["Fare"].mean() - data["Fare"].std() * factor 
##평균에서 표준편차를 더하고 빼서 비교하는 것은 평균에서 너무 멀어진 값 처리

data = data[(data["Fare"] < max_) & (data["Fare"] > min_)] ##& (AND) |(OR)

display(data)
```

```python
## 원 핫 인코딩

data = pd.read_csv("titanic.csv")

encoded1 = pd.get_dummies(data["Sex"])
encoded2 = pd.get_dummies(data["Pclass"], "class")

display(encoded2)
```

```python
## 스케일링
data = pd.read_csv("titanic.csv")

##나이값 결측치 평균
data = data[["Age", "Fare"]]
# 이건 안됨
# data = data["Age", "Fare"] # 두 개 이상 컬럼을 부를 때 대괄호 두개를 써야함

data = data.fillna(data.mean())

original_value = data["Age"] # 원래 값
min_ = data["Age"].min() # min() 최솟값 부르는 함수
max_ = data["Age"].max() # max() 최댓값 부르는 함수
data["Age"] = (original_value - min_) / (max_ - min_) ## 1과 0 사이로 바뀜 - 스케일링
data["Fare"] = (data["Fare"] - data["Fare"].mean()) / (data["Fare"].std()) # std: 표준편차

display(data)
```
