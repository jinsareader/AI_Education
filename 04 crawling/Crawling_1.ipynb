#### String , text 차이
- string : 태그 하위에 있는 문자열을 객체화시켜서 문자열이 없으면 None을 반환
- text: 하위자식태그의 텍스트까지 문자열을 반환

```python
html = """
<html>
<body>
  <h1>글자크기</h1>
  <h2>글자크기2</h2>
  <p>원하는 글자 추출</p>
  <p>크롤링 수업시간 </p>
  <p></p>
  <ul>
      <li><a href="http://www.daum.net">daum</a></li>
      <li><a href="http://www.naver.com">naver</a></li>
  </ul>
</body>
</html>  """

soup = BeautifulSoup(html,"lxml")
txt = soup.find("a")
txt.get_text()
print(txt.string);
print(txt.text);
```

#### 원하는 요소에 접근하느 메소드
- find(tag) : 원하는 태그 하나만 반환
- find_all(tag) : 원하는 태그를 리스트형태로 반환

```python
txt = soup.find_all("p")
print(txt);

for t in txt :
    print(t.string);
for t in txt :    
    print(t.text);
```

```python
txt = soup.find_all("a")

for t in txt :
    print(t.text);
```

- 태그의 속성값 추출 : .attrs['속성']

```python
text = soup.find_all("a");
for t in text :
    print(t.attrs['href']);
```

- id : 문서안에서 한번만 사용
- find(tag,id=id값)
- class : 공통적으로 적용
- find(tag, class_ = class값)
- find(tag, class값)

```python
html="""
<html>
<body>
    <h1 id="title">글자크기</h1>
    <p class="cstyle">원하는 글자 추출</p>
    <p id="body" align="center"> 크롤링 수업 시간</p>
</body>
</html>
"""

soup = BeautifulSoup(html,"lxml");
text = soup.find("h1", id = "title")

print(text)
```

```python
text = soup.find("p",id = "body")
print(text.text);
```

```python
text = soup.find("p", class_ = "cstyle")
print(text.text);
```

'원하는 글자 추출'

- 태그의 속성값 : .find(tag,attrs={'속성명':'속성값'})

```python
data = soup.find('p',attrs={'align':'center'})
print(data.text);
```

#### 실습2. 뉴스제목 가져오기
- h3태그에 클래스값이 "tit_view" 이용하여 텍스트 가져오기

```python
res = requests.get("https://v.daum.net/v/20240405104907867");
soup = BeautifulSoup(res.content,"lxml")
data = soup.find("h3", class_ = "tit_view")
print(data.text);
```

```python
res = requests.get("https://v.daum.net/v/20240405104907867");
soup = BeautifulSoup(res.content,'lxml')
title = soup.find("h3", class_ = "tit_view")
content = soup.find("div", class_ = "article_view")
print(title.text)
print(content.text.strip())
```

```python
res = requests.get("https://v.daum.net/v/20240405104907867");
soup = BeautifulSoup(res.content,'lxml')
title = soup.find_all("span", class_ = "txt_info")
for s in title :
    print(s.text)
```
