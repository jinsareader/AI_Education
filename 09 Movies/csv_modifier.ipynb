```python
import pandas as pd
import custom
from IPython.display import display
```

```python
df = pd.read_csv("data/critic_reviews.csv")
#필요없는 열 삭제
df = df[["quote","scoreSentiment"]]
display(df)
```

```python
#결측치 삭제
df = df.drop(df[df.isnull().sum(axis = 1) > 0].index, axis = 0 )
#POSITIVE, NEGATIVE 숫자로(t)
df = df.replace({"scoreSentiment" : {"POSITIVE" : 1}})
df = df.replace({"scoreSentiment" : {"NEGATIVE" : 0}})
#문장 전처리
for i in range(len(df)) :
    df.iloc[i, 0] = custom.text_preprocess(df.iloc[i, 0])
#단어길이열 추가 (padding 작업용)
word_len = []
for s in df["quote"].values.tolist() :
    word_len.append(len(s.split()))
df["word_len"] = word_len
#출력
display(df)



```

```python
df.to_csv("data/movie_data.csv")
```

```python
#단어 10개 초과 제거 
df = df.drop(df[df["word_len"] > 10].index, axis = 0)
df.to_csv("data/simple_movie_data.csv")
```
