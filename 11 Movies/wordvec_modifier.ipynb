```python
import numpy
import custom
import pickle
```

```python
with open("data/model.txt", mode = "r") as f:
    original_data = f.read()
```

```python
data = original_data.replace("_"," ")
data = data.split("\n")

vec_dict = {}
line_len = len(data[1].split())
vec_dict["<pad>"] = list(numpy.zeros(shape = (line_len - 2)))
vec_dict["<unk>"] = list(numpy.zeros(shape = (line_len - 2)))

for i in range(1,len(data)) :
    vec = []
    line = data[i].split()
    if len(line) != line_len : #벡터 길이 맞지 않으면 넘김
        continue
    word = custom.text_preprocess(line[0]).replace(" ","")
    if word == "" or word in vec_dict : #빈 단어면 넘김 or 이미 있는 단어면 넘김
        continue
    for j in range(2,len(line)) :
        vec.append(float(line[j]))
    vec_dict[word] = vec

print(len(vec_dict))
```

```python
test = list(vec_dict.values())[2]
print(test)
print(type(test))
print(len(test))
```

```python
with open("data/vector.pkl", mode = "wb") as f :
    pickle.dump(vec_dict, f)

```
