```python
import sys
import os
sys.path.append(os.path.join(os.path.dirname(""), '..'))

import pandas as pd
import custom
from IPython.display import display
from tqdm import tqdm
```

```python
df = pd.read_csv("data/critic_reviews.csv")
#필요없는 열 삭제
df = df[["quote","scoreSentiment"]]
display(df)
```

```python
#결측치 삭제
df = df.drop(df[df.isnull().sum(axis = 1) > 0].index, axis = 0 )
#POSITIVE, NEGATIVE 숫자로(t)
df = df.replace({"scoreSentiment" : {"POSITIVE" : 1}})
df = df.replace({"scoreSentiment" : {"NEGATIVE" : 0}})
#문장 전처리
for i in tqdm(range(len(df))) :
    df.iloc[i, 0] = custom.text_preprocess(df.iloc[i, 0].replace("n't"," not"))
#단어길이열 추가 (padding 작업용)
word_len = []
for s in df["quote"].values.tolist() :
    word_len.append(len(s.split()))
df["word_len"] = word_len

df = df.drop(df[df["word_len"] == 0].index, axis = 0 )
#출력
display(df)



```

```python
df.to_csv("data/movie_data.csv", index = False)
```

```python
#많은 단어 제거
#max_len = df["word_len"].mean() + df["word_len"].std()
max_len = 10
simple_df = df.drop(df[df["word_len"] > max_len].index, axis = 0)

print(max_len)
display(simple_df)
```
