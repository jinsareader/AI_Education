```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# 주소:https://finance.naver.com/sise/sise_quant.naver?sosok=0
```

```python
# 사이트 접근하기
res = req.get("https://finance.naver.com/sise/sise_quant.naver?sosok=0")
# html로 변경하기
data = bs(res.content,"html");
```

```python
# 종목명 받아오기
table = data.select_one("table.type_2")
row = table.select("tr > td:nth-of-type(2)");
title_list = []
for i in range(0,len(row)) :
    title = row[i].text
    title_list.append(title)

for t in title_list:
    print(t);
```

```python
# 현재가 받아오기
row = table.select("tr > td:nth-of-type(3)");
price_list = []
for i in range(0,len(row)) :
    price = row[i].text;
    price_list.append(price)

for p in price_list:
    print(p);
```

```python
# 거래량 받아오기
row = table.select("tr > td:nth-of-type(6)");
trade_list = []
for i in range(0,len(row)) :
    trade = row[i].text;
    trade_list.append(trade)

for trade in trade_list:
    print(trade);
```

```python
#딕셔너리로 합쳐주기
dic = {'종목명' : title_list, '현재가' : price_list, '거래량' : trade_list}
```

```python
# 데이터프레임으로 변환하기
dt = pd.DataFrame(dic);
dt.head(10)
```

```python
# 230426.csv  파일 만들기
with open("230426.csv","w",encoding="utf-8") as file :
    lenth = min(len(dic['종목명']),len(dic['현재가']),len(dic['거래량']))
    file.write("종목명,현재가,거래량\n");
    for i in range(0,lenth) :
        file.write(dic['종목명'][i]+','+dic['현재가'][i].replace(',','')+','+dic['거래량'][i].replace(',',''));
        file.write("\n");
```

```python
dt.to_csv("230426.csv")
```

### iframe 내용 가져오기 : 문서내에 다른 문서가 있을 때

```python
url_main = "https://finance.naver.com/"
url_sub = "marketindex/"
page = url_main+url_sub
res = req.get(page)
soup = bs(res.content,"lxml")
```

```python
frameSrc=soup.find('iframe',id='frame_ex1')
print(frameSrc.attrs['src'])
```

```python
url = url_main+frameSrc.attrs['src'];
res = req.get(url);
soup = bs(res.content,'lxml');

table = soup.select_one("table.tbl_exchange");

row = table.select("tr > td:nth-of-type(1)");
name_list = [];
for r in row :
    name_list.append(r.text.strip());

row = table.select("tr > td:nth-of-type(2)");
exchange_list = [];
for r in row :
    exchange_list.append(r.text);

length = min(len(name_list),len(exchange_list));
dic = {"순번":range(1,length+1),"통화명":name_list[:length],"매매기준율":exchange_list[:length]}
df = pd.DataFrame(dic,columns=["순번","통화명","매매기준율"]);
df.set_index("순번",inplace=True)
print(df.head(10))
df.to_csv("money.csv")
```
