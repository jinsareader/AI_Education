```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
res = requests.get("https://n.news.naver.com/article/374/0000436280?cds=news_media_pc&type=editn")
```

```python
print(res.status_code)
#200 정상
#4XX 클라이언트 오류
#5XX 서버 오류
```

```python
print(res.content)
```

```python
soup = bs(res.content, "lxml")
print(soup)
```

```python
article = soup.find("article")

print(article)
```

```python
print(article.text, end = "\n\n") #모든 자식의 텍스트 출력
```

```python
print(article.get_text(strip = "True", separator = "\n"), end = "\n\n") #모든 자식의 텍스트 출력, strip, saparator 지정 가능
```

```python
print(article.string, end = "\n\n") #해당 HTML 태그 안에 하나의 자식만 있을 때에만 텍스트 출력 
```
