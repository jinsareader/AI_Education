```python
import numpy
import matplotlib.pyplot as plt
from scipy.io import wavfile
from python_speech_features import mfcc, logfbank
```

```python
sampling_freq, signal = wavfile.read("data/banana/banana03.wav")
```

```python
print("sampling Hz : ", sampling_freq)
print("signal_shape :", signal.shape)
print("datatype :", signal.dtype)
print("signal duration:", round(signal.shape[0] / float(sampling_freq), 2))
```

```python
features_mfcc = mfcc(signal, sampling_freq)

print("Number of windows =", features_mfcc.shape[0])
print("Length of each feature =", features_mfcc.shape[1])
print("Type of Data =", features_mfcc.dtype)
```

```python
plt.matshow(features_mfcc)
plt.title("MFCC")
```

```python
features_fb = logfbank(signal, sampling_freq)

print("Number of windows =", features_fb.shape[0])
print("Length of each feature =", features_fb.shape[1])
print("Type of Data =", features_mfcc.dtype)
```

```python
plt.matshow(features_fb)
plt.title("Filter bank")
```
