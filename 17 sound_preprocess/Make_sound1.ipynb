```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
```

```python
output_file = "generated_audio.wav"
```

```python
duration = 4 #4ì´ˆ
sampling_freq = 44100 #Hz
tone_freq = 784
min_val = -4 * np.pi
max_val = 4 * np.pi
```

```python
t = np.linspace(min_val, max_val, duration * sampling_freq)
signal = np.sin(2 * np.pi * tone_freq * t)
```

```python
noise = 0.5 * np.random.rand(duration * sampling_freq)
signal += noise
```

```python
scaling_factor = (2**15) - 1
signal_normalize = signal / np.max(np.abs(signal))
signal_scaled = np.int16(signal_normalize * scaling_factor)
```

```python
write(output_file, sampling_freq, signal_scaled)
```

```python
sample_signal = signal[:500]

time_axis = 1000 * np.arange(0, len(sample_signal), 1) / float(sampling_freq)
```

```python
plt.plot(time_axis, sample_signal, color = "black")
plt.xlabel("Time (milisecs)")
plt.ylabel("Amplitude")
plt.show()
```
