```python
import numpy
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
sampling_freq, signal = wavfile.read("sample/spoken_word.wav")
```

```python
print("sampling Hz : ", sampling_freq)
print("signal_shape :", signal.shape)
print("datatype :", signal.dtype)
print("signal duration:", round(signal.shape[0] / float(sampling_freq), 2))
```

```python
signal = signal / (2**15)
# -(2^15) ~ (2^15)-1
```

```python
time_axis = 1000 * numpy.arange(0, len(signal), 1) / float(sampling_freq)
# 1000을 곱한 이유는 milisecs
```

```python
fig, ax = plt.subplots()
ax.plot(time_axis, signal, color = "black")
plt.xlabel("Time (milisecs)")
plt.ylabel("Aplitude")

plt.show()
```
