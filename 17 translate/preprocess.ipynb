```python
import sys
import os
sys.path.append(os.path.join(os.path.dirname(""),".."))
import custom
import numpy as np

import pandas as pd
from kiwipiepy import Kiwi
from tqdm import tqdm
import json
```

```python
with open("data/eng_kor.txt", mode = "r", encoding="UTF8") as f :
    data = f.readlines()

print(len(data))
```

```python
eng = []
eng_len = []
kor = []
kor_len = []

for d in tqdm(data) :
    d = d.split("\t")
    e = custom.text_preprocess(d[0], True)
    e = e.replace("."," .")
    e = e.replace("!"," !")
    e = e.replace("?"," ?")
    k = custom.text_preprocess_kor(d[1], True)
    
    eng.append(e)
    eng_len.append(len(e.split()))
    kor.append(k)
    kor_len.append(0)

df = pd.DataFrame({'eng' : eng, 'eng_len' : eng_len, 'kor' : kor, 'kor_len' : kor_len})

display(df)
```

```python
kiwi = Kiwi()

for i in tqdm(range(len(df))) :
    words = []
    
    temp = df.iloc[i,2]
    temp = kiwi.tokenize(temp, z_coda=True, split_complex=True, compatible_jamo=True)
    for w in temp :
        words.append(w.form)
    words.append("<eos>")
    df.iloc[i,2] = " ".join(words)
    df.iloc[i,3] = len(words)

display(df)
```

```python
print(df['eng_len'].mean())
print(df['eng_len'].std())
print()
print(df['kor_len'].mean())
print(df['kor_len'].std())

```

```python
p = len(df[(df['eng_len'] <= 10) & (df['kor_len'] <= 20)]) / len(df)

print(p)

new_df = df[(df['eng_len'] <= 10) & (df['kor_len'] <= 20)]
display(new_df)
```

```python
new_df.to_csv("data/data.csv", index=False, encoding="UTF8", sep = "\t")
```

```python
eng_word_dict, eng_num_dict  = custom.make_dict(new_df['eng'].values.tolist(), special_words=['<pad>','<unk>','<sos>','<eos>'])

kor_word_dict, kor_num_dict  = custom.make_dict(new_df['kor'].values.tolist(), special_words=['<pad>','<unk>','<sos>','<eos>'])

print(len(eng_word_dict))
print(len(kor_word_dict))
```

```python
with open('data/eng_word.json', mode = "w", encoding="UTF8") as f:
    json.dump(eng_word_dict, f)

with open('data/eng_num.json', mode = "w", encoding="UTF8") as f:
    json.dump(eng_num_dict, f)

with open('data/kor_word.json', mode = "w", encoding="UTF8") as f:
    json.dump(kor_word_dict, f)

with open('data/kor_num.json', mode = "w", encoding="UTF8") as f:
    json.dump(kor_num_dict, f)
```
