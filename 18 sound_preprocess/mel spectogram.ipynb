```python
import os
import IPython
os.environ['NUMBA_CACHE_DIR'] = IPython.paths.get_ipython_cache_dir()

import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
sig, freq = librosa.load("sample/orange01.wav", sr = 16000)
S = librosa.feature.melspectrogram(y=sig, sr=freq, n_mels=128, fmax = 8000)
```

```python
plt.matshow(S)
plt.title("mel spectogram")
```
